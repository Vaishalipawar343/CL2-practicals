```python
dna_seq = "ATGCGATACGTAGCTAGCGTAGCTAAATGCGGCTAGCTGA"

def gc_content(seq):
    g = seq.count("G")
    c = seq.count("C")
    return (g + c) / len(seq) * 100

def find_motif(seq, motif):
    positions = []
    for i in range(len(seq) - len(motif) + 1):
        if seq[i:i+len(motif)] == motif:
            positions.append(i + 1)  # 1-based indexing
    return positions

def find_orfs(seq):
    start_codon = "ATG"
    stop_codons = ["TAA", "TAG", "TGA"]
    orfs = []
    for i in range(len(seq) - 2):
        codon = seq[i:i+3]
        if codon == start_codon:
            for j in range(i + 3, len(seq) - 2, 3):
                stop_codon = seq[j:j+3]
                if stop_codon in stop_codons:
                    orfs.append(seq[i:j+3])
                    break
    return orfs

print("DNA Sequence:", dna_seq)
print("GC Content (%):", gc_content(dna_seq))
print("Motif 'ATG' found at positions:", find_motif(dna_seq, "ATG"))
print("Identified ORFs:", find_orfs(dna_seq))
```
